```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
```

### Read in the cleaned-up text file. 
Each line is a Q or A, and is treated as a document  


```python
filename = "web_questions.txt"

with open(filename, "r+", encoding = "utf-8") as f:
	lines = f.readlines()
	print(len(lines))
```

```python
# create English stop words list and remove stop words from tokens
en_stop = get_stop_words('en')
print(en_stop)
```

### Creat stopped and stemmed tokens

```python
texts = []

for line in lines: 
	# tokenization
	tokenizer = RegexpTokenizer(r'\w+')

	raw = line.lower()
	print(raw)

	tokens = tokenizer.tokenize(raw)
	print(tokens)

	stopped_tokens = [t for t in tokens if not t in en_stop]
	print(stopped_tokens)

	# Create p_stemmer of class PorterStemmer and creat stemmed tokens 
	p_stemmer = PorterStemmer()
	stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
	print(stemmed_tokens)

	texts.append(stemmed_tokens)
```

### Constructing a document-term matrix
The Dictionary() function traverses texts, assigning a unique integer id to each unique token while also collecting word counts and relevant statistics. To see each token’s unique integer id, try print(dictionary.token2id).

```python
dictionary = corpora.Dictionary(texts)
# print(dictionary.token2id)

# doc2bow() function converts dictionary into a bag-of-word
corpus = [dictionary.doc2bow(text) for text in texts]

print(corpus)
```

### Generating LDA model

```python
# generate an LDA model
ldamodel = models.ldamodel.LdaModel(corpus, num_topics=7, id2word = dictionary, passes=20)

# review topics 
print(ldamodel.print_topics(num_topics=10, num_words=15))
```
